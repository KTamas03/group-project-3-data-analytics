```python
# Import dependencies
import pandas as pd
from pathlib import Path
```

```python
# Files to Load
accident_path = Path("../Cleaned_Datasets/accident_cleaned.csv")
node_path = Path("../Cleaned_Datasets/node_cleaned.csv")
atmos_path = Path("../Cleaned_Datasets/atmospheric_cond_cleaned.csv")
surface_path = Path("../Cleaned_Datasets/road_surface_cond_cleaned.csv")
```

```python
# Read the CSV file
accident_data = pd.read_csv(accident_path, low_memory=False)
node_data = pd.read_csv(node_path, low_memory=False)
atmos_data = pd.read_csv(atmos_path, low_memory=False)
surface_data = pd.read_csv(surface_path, low_memory=False)
```

```python
# show accident df
accident_data_df = pd.DataFrame(accident_data)
accident_data_df.head()
```

```python
# List the columns
accident_data_df.dtypes
```

```python
# show node df
node_data_df = pd.DataFrame(node_data)
node_data_df.head()
```

```python
# show atmos df
atmos_data_df = pd.DataFrame(atmos_data)
atmos_data_df.head()
```

```python
# show surface df
surface_data_df = pd.DataFrame(surface_data)
surface_data_df.head()
```

```python
# merge accident and node data
merge1_df = pd.merge(accident_data_df, node_data_df, on='ACCIDENT_NO', how='left')
merge1_df.head()
```

```python
# merge accident, node and atmos data
merge2_df = pd.merge(merge1_df, atmos_data_df, on='ACCIDENT_NO', how='left')
merge2_df.head()
```

```python
# merge accident, node, atmos and surface data
merge3_df = pd.merge(merge2_df, surface_data_df, on='ACCIDENT_NO', how='left')
merge3_df
```

```python
# List the columns
merge3_df.dtypes
```

```python
# Create a dictionary to specify the column name changes
new_column_names = {'Accident Type Desc': 'Accident_Type_Desc',
                    'Day Week Description': 'Day_Week_Description',
                    'DCA Description': 'DCA_Description',
                    'Light Condition Desc ': 'Light_Condition_Desc ',
                    'Road Geometry Desc ': 'Road_Geometry_Desc ',
                   }

# Use the rename method to rename columns
merge3_df.rename(columns=new_column_names, inplace=True)

# Display the DataFrame with renamed columns
merge3_df
```

```python
# Drop rows with NaN values in 'LGA_NAME'
column_name = 'LGA_NAME' 

merge3_df = merge3_df.dropna(subset=[column_name])

# Display the DataFrame after removing rows with NaN 
merge3_df
```

```python
# Storing file in Data folder
merged_data_output_path = Path("../Cleaned_Datasets/master_accident_cleaned.csv")
merge3_df.to_csv(merged_data_output_path, index=None, header=True)
```
